```python
%%bash

# Import dependencies 
# gensim library and tools for implementing Word2Vec
import gensim
from gensim.models import Word2Vec
from gensim.models import FastText

import argparse

import pickle


# Command Line Arguments
parser = argparse.ArgumentParser()

parser_model = parser.add_argument_group('model options')
parser_model.add_argument('--model',choices=['w2v','fast'])
parser_model.add_argument('--data',choices=['latin', 'greek', 'arabic', 'bengali', 'old_english', 'middle_english', 'french', 'hebrew', 'hindi', 'marathi', 'old_norse', 'sanskrit', 'punjabi', 'telugu', 'middle_high_german', 'gujarati', 'malayalam', 'old_swedish'])

parser_tune = parser.add_argument_group('model tuning hyperparameters')
parser_tune.add_argument('--architecture',choices=['cbow','sg'])
parser_tune.add_argument('--min_count',type=int)
parser_tune.add_argument('--size',type=int)
parser_tune.add_argument('--window',type=int)
parser_tune.add_argument('--epochs',type=int,default=2)
parser_tune.add_argument('--lr_exp',type=float)
parser_tune.add_argument('--lemmatization',action='store_true')



args = parser.parse_args()

####################
### Train the Model
####################

# unpickle books

lem_flag =''
if args.lemmatization:
    lem_flag='-lemmed'
    with open('./corpora/preprocessed/'+args.data+'_txt-lemmed.p','rb') as p:
        data = pickle.load(p)
else:      
    with open('./corpora/preprocessed/'+args.data+'_txt-.p','rb') as p:
        data = pickle.load(p)


lr= 10**args.lr_exp
sg=0
if args.architecture == 'sg':
    sg=1
if args.model == 'w2v':
    f = Word2Vec

if args.model == 'fast':
    f = FastText

model = f(sentences=data, size=args.size, window=args.window, alpha=lr, sg=sg, min_count=args.min_count, iter=2, seed=47, workers=1)

model.save('models/'+args.data+'-'+args.model+'-s'+str(args.size)+'-w'+str(args.window)+'-lr=1e'+str(args.lr_exp)+'-min'+str(args.min_count)+'-'+args.architecture+lem_flag+'.model')
```
